```python
import numpy as np
import pandas as pd
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv("users.data",sep = '\t',names = column_names)
df.head()
```

```python
len(df)
```

```python
movie_titles = pd.read_csv("movie_id_titles.csv")
movie_titles.head(20)
```

```python
len(movie_titles)
```

```python
df = pd.merge(df, movie_titles, on='item_id')
df.head()
```

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
type(moviemat)
```

```python
def recommend():
  film = input("Film adı giriniz: ")
  movie_user_ratings = moviemat[film]
  movie_user_ratings.head()
  ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
  ratings['rating_oy_sayisi'] = pd.DataFrame(df.groupby('title')['rating'].count())
  similar_to_movie = moviemat.corrwith(movie_user_ratings)
  corr_starwars = pd.DataFrame(similar_to_movie, columns=['Correlation'])
  corr_starwars = corr_starwars.join(ratings['rating_oy_sayisi'])
  #corr_starwars[corr_starwars['rating_oy_sayisi']>100].sort_values('Correlation',ascending=False).head()
  result = corr_starwars[corr_starwars['rating_oy_sayisi']>100].sort_values('Correlation',ascending=False)
  print(result.iloc[[1,2,3],[0]])

recommend()
```
